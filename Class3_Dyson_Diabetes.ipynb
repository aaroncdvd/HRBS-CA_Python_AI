#[>](https://www.kaggle.com/uciml/pima-indians-diabetes-database/data) Sample Dataset 1 : Diabetes

**Context**
This dataset is originally from the National Institute of Diabetes and Digestive and Kidney Diseases. The objective of the dataset is to diagnostically predict whether or not a patient has diabetes, based on certain diagnostic measurements included in the dataset. Several constraints were placed on the selection of these instances from a larger database. In particular, all patients here are females at least 21 years old of Pima Indian heritage.

**Content**
The datasets consists of several medical predictor variables and one target variable, Outcome. Predictor variables includes the number of pregnancies the patient has had, their BMI, insulin level, age, and so on.

###Import CSV Data

```python
import pandas as pd # Import Pandas Module
```

```python
diabetes = pd.read_csv( # Load the diabetes dataset from CSV
      'https://raw.githubusercontent.com/plotly/datasets/master/diabetes.csv',header=0)
```

```python
diabetes # Show diabetes data
```

###Histograms

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20,20)) # Set Default Figure Size
plt.rcParams.update({'font.size': 22}) # Set Default Font Size
```

####*One Variable*

```python
Glucose = diabetes[['Glucose']] # Take Glucose column from Diabetes
```

```python
Glucose.plot.hist(figsize=(20,10),bins=20,fontsize=20) # Graph Glucose as Histogram
```

####*One Variable (Colored By Another Variable)*

```python
outcomes = diabetes.Outcome.unique() # Get List of Outcomes from diabetes Dataset
plt.figure(figsize=(20,10))
plt.hist([diabetes.loc[diabetes.Outcome == x, 'Glucose'] for x in outcomes], label=outcomes,stacked=True, bins=20)
plt.legend(outcomes) # Plot Outcomes in Sepatate 'Stacked' Histograms 
```

####*Two Variables*

```python
GlucoseAndBP = diabetes[['Glucose', 'BloodPressure']]
GlucoseAndBP.plot.hist(figsize=(20,10), bins=20, alpha=0.5) # Plot Glucose and BloodPressure Columns as Histograms
```

####*All Variables*

```python
diabetes.hist(bins=30, figsize=(30, 20)) # Plot Histograms for Every Column
```

###Transform

```python
diabetes['Insulin']= (diabetes['Insulin'] / 7) # Scale Columns By Linear Transformations to Improve Scatterplot and Boxplot
diabetes['DiabetesPedigreeFunction']= (diabetes['DiabetesPedigreeFunction'] * 40)
diabetes['Pregnancies']= (diabetes['Pregnancies'] * 5)
diabetes['Glucose']= (diabetes['Glucose'] / 3)
diabetes['BMI']= (diabetes['BMI'] / 1)
diabetes['Outcome']= (diabetes['Outcome'] * 1)
```

###Boxplot

```python
diabetes.boxplot(figsize=(25,10),fontsize=20 # Create Dataset Boxplot
              ,grid=False,flierprops=dict(marker='o', markerfacecolor='g', markersize=8, linestyle='none', markeredgecolor='g'))
```

###Scatterplot by Size

```python
diabetes.plot.scatter(x='Glucose', # Create Scatterplot By Size
                      y='BloodPressure',
                      s='Insulin',
                      figsize=(25, 10))
```

###Scatterplot by Color

```python
fig, ax = plt.subplots() # WATCH OUT BUG! https://stackoverflow.com/questions/43121584/matplotlib-scatterplot-x-axis-labels
diabetes.plot.scatter(x='Insulin', # Create Scatterplot By Color
                      y='Glucose',
                      c='Outcome',
                      colormap='viridis',
                      figsize=(25, 10),ax=ax)
```

###Correlations

```python
correlations=diabetes.corr() # Generate Dataset Correlation Matrix
```

```python
correlations # Show correlations
```

```python
import seaborn as sn # Import Module Seaborn
```

```python
plt.subplots(figsize=(25,10)) # Generate Dataset Correlations Heatmap:
sn.diverging_palette(250, 30, l=65, center="dark", as_cmap=True) # Color : Fire is HOT, Ice is COLD
sn.heatmap(correlations, annot=False,vmin=-1, vmax=1,center=0)
plt.xticks(rotation=43)
```

###**Homework: Did our linear transformations affect the correlations?**


**NO** 
Linear transformations have no effect on Pearson's correlation coefficient. But, they are tools to help us for interpretation and visualization!

**EXAMPLE:** the correlation between height and weight is the same regardless of whether height is measured in inches, feet, centimeters or even miles.
